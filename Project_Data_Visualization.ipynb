# To Display Value Count For Two Categorical Variables

# Importing The Required Library

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Load The Data

```python
df = pd.read_csv("E:\Project_Work\Test_data.csv")
```

# Count Plot On Two Categorical Variables

```python
sns.countplot(x ='gender', hue = "department", data = df)
```

# Count Plot On Two Categorical Variables

```python
sns.countplot(x ='gender', hue = "education", data = df)
```

# To Display A Boxplot Using Only One Variable

```python
sns.set(style='whitegrid')
sns.boxplot(x='age', data=df)
```

# To Display A Boxplot Using Only One Variable

```python
sns.set(style='whitegrid')
sns.boxplot(x='avg_training_score', data=df)
```
